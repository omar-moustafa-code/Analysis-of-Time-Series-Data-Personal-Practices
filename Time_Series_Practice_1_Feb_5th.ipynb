```r
# @title Time Series Practice 1 - Feb. 5th
```

```r
# @title Load & View the Data
```

```r
# ‘scan’ is the same as read.csv, read.table, read.excel, etc., but it is used for UNIVARIATE data

df = scan("nybirths.dat")
```

```r
head(df)
```

```r
# @title Convert the Data to Time Series
```

```r
# ‘ts’ stands for time series

# frequency = 12 because it is a data set of the number of births PER MONTH (MONTHLY DATA)

# ‘start’ refers to the starting point of the data collection where 1948 is the starting year and 1 is the starting month (which is January)

df_ts = ts(df, frequency = 12, start = c(1948, 1))
```

```r
# @title Plot the Time Series Data
```

```r
plot.ts(df_ts)
```
